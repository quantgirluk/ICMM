## Financial Data in Python

```python
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.figure_format ='retina'
sns.set(style="whitegrid")
```

### Equities

[TSLA in Yahoo Finance Website](https://finance.yahoo.com/quote/TSLA/)

```python
tesla_df = yf.download('TSLA')
tesla_df.head()
```

```python
tesla_df.tail()
```

```python
plt.figure(figsize=(12, 6))
tesla_df['Close'].plot(color='blue', title="TESLA Close Prices")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
mask = tesla_df.index > '2020-01-01'
tesla_df[mask]['Close'].plot(color='blue',  title="TESLA Close Prices")
plt.show()
```

### Commodities

```python
crude_oil_df = yf.download('CL=F')
plt.figure(figsize=(12, 6))
crude_oil_df ['Close'].plot(color='purple', title="Crude Oil WTI Prices")
plt.show()
```

```python
gold_df = yf.download('GC=F')
plt.figure(figsize=(12, 6))
gold_df ['Close'].plot(color='orange', title="Gold COMEX XAU/USD")
plt.show()
```

### Interest Rates

```python
btc_df = yf.download('^FVX')
plt.figure(figsize=(12, 6))
btc_df['Close'].plot(color='green', title="Treasury Yield 5 Years")
plt.show()
```

To get a better understanding of yield curves and its term structure I always recommend looking at this [visualisation by NYT](https://www.nytimes.com/interactive/2015/03/19/upshot/3d-yield-curve-economic-growth.html)

### Crypto

```python
btc_df = yf.download('BTC-USD')
plt.figure(figsize=(12, 6))
btc_df['Close'].plot(color='red', title="Bitcoin Price USD")
plt.show()
```
